```python
from ultralytics import YOLO
import torch
import os

dataset_path = os.getcwd() + '\\Datasets'
# print(dataset_path)

model = YOLO("yolo26n-cls.pt")
```

```python
results = model.train(
    data=dataset_path,
    epochs=50,
    imgsz=224,
    batch=16,
    device="cpu",
    project="Fruit-Vision",
    name="yolo26_fruit_run"
)
```

```python
best_model = YOLO(os.getcwd() + "/runs/classify/Fruit-Vision/yolo26_fruit_run/weights/best.pt")

metrics = best_model.val(data=dataset_path)

print(f"Top-1 Accuracy: {metrics.top1:.4f}")
print(f"Top-5 Accuracy: {metrics.top5:.4f}")
```

```python
class_names = ['freshapples', 'freshbanana', 'freshoranges', 'rottenapples', 'rottenbanana', 'rottenoranges']

images = ['freshoranges/rotated_by_15_Screen Shot 2018-06-12 at 11.50.47 PM.png',
          'rottenbanana/rotated_by_15_Screen Shot 2018-06-12 at 8.48.18 PM.png',
          'rottenapples/rotated_by_15_Screen Shot 2018-06-07 at 2.18.25 PM.png',
          "freshbanana/rotated_by_15_Screen Shot 2018-06-12 at 9.38.29 PM.png"
         ]

for image in images:
    image_path =  os.getcwd() + f"/datasets/test/{image}"
    results = best_model(image_path)
    probs = results[0].probs
    pred_idx = probs.top1
    confidence = probs.top1conf.item()
    print(f"Predicted class: {class_names[pred_idx]}")
    print(f"Confidence: {confidence:.4f}")
```

```python
best_model.export(format="onnx")
```

```python
def detect(image_path: str):
    results = best_model(image_path)
    probs = results[0].probs
    pred_idx = probs.top1
    confidence = probs.top1conf.item()
    print(f"Predicted class: {class_names[pred_idx]}")
    print(f"Confidence: {confidence:.4f}")

detect(os.getcwd() + "/test/freshBanana.jpg")
detect(os.getcwd() + "/test/Banana.jpeg")
detect(os.getcwd() + "/test/images.jpeg")
detect(os.getcwd() + "/test/images (1).jpeg")
```
